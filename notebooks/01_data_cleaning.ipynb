## 1. Setup & Import Libraries

```
import pandas as pd
import string
import re

# For Google Colab - upload files
# from google.colab import files
# uploaded = files.upload()

print("Libraries imported successfully!")
```

## 2. Load Glassdoor Data

```
# Load Glassdoor reviews
glassdoor_df = pd.read_csv('glass_reviews_raw.csv')

print(f"Shape: {glassdoor_df.shape}")
print(f"\nColumns: {glassdoor_df.columns.tolist()}")
glassdoor_df.head()
```

## 3. Text Cleaning Function

We create a reusable function that:
- Converts to lowercase
- Removes punctuation
- Removes short words (< 3 characters)
- Strips extra whitespace


```
def clean_text(text):
    """Clean text for NLP analysis"""
    if pd.isna(text):
        return ""
    
    # Convert to lowercase
    text = str(text).lower()
    
    # Remove punctuation
    text = text.translate(str.maketrans("", "", string.punctuation))
    
    # Split into words and remove short words
    tokens = text.split()
    tokens = [word for word in tokens if len(word) > 2]
    
    # Rejoin
    return " ".join(tokens)

# Test the function
test_text = "This is a TEST! With punctuation... and SHORT words."
print(f"Original: {test_text}")
print(f"Cleaned: {clean_text(test_text)}")
```

## 4. Clean Glassdoor Reviews

We'll clean both the `review_pros` and `review_cons` columns.


```
# Clean PROS column
glassdoor_df['pros_cleaned'] = glassdoor_df['review_pros'].apply(clean_text)

# Clean CONS column  
glassdoor_df['cons_cleaned'] = glassdoor_df['review_cons'].apply(clean_text)

# Create combined cleaned text for overall analysis
glassdoor_df['cleaned_text'] = glassdoor_df['pros_cleaned'] + " " + glassdoor_df['cons_cleaned']

# Preview results
glassdoor_df[['review_pros', 'pros_cleaned', 'review_cons', 'cons_cleaned']].head()
```

## 5. Load and Clean YouTube Transcripts

```
# Load YouTube transcripts
youtube_df = pd.read_csv('youtube_transcripts_raw.csv')

print(f"Shape: {youtube_df.shape}")
print(f"\nColumns: {youtube_df.columns.tolist()}")
youtube_df.head()
```

```
# Clean transcript column
youtube_df['transcript_cleaned'] = youtube_df['transcript'].apply(clean_text)

# Preview
youtube_df[['Video Title', 'transcript', 'transcript_cleaned']].head()
```

## 6. Data Quality Check

```
# Check for empty cleaned text
print("=== Glassdoor Data Quality ===")
print(f"Total rows: {len(glassdoor_df)}")
print(f"Empty cleaned_text: {(glassdoor_df['cleaned_text'] == '').sum()}")
print(f"Null values in cleaned_text: {glassdoor_df['cleaned_text'].isna().sum()}")

print("\n=== YouTube Data Quality ===")
print(f"Total rows: {len(youtube_df)}")
print(f"Empty transcript_cleaned: {(youtube_df['transcript_cleaned'] == '').sum()}")
print(f"Null values in transcript_cleaned: {youtube_df['transcript_cleaned'].isna().sum()}")
```

## 7. Save Cleaned Data

```
# Save cleaned Glassdoor data
glassdoor_df.to_csv('glassdoor_cleaned.csv', index=False)
print("✅ Saved: glassdoor_cleaned.csv")

# Save cleaned YouTube data
youtube_df.to_csv('youtube_cleaned.csv', index=False)
print("✅ Saved: youtube_cleaned.csv")
```
